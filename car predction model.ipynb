```python
import pandas as pd 
from matplotlib import pyplot as plt
df = pd.read_csv('car data.csv')
df.head()
print(df['Year'].unique())
print(df['Seller_Type'].unique())
print(df['Fuel_Type'].unique())
print(df['Owner'].unique())
df.isnull().sum()
df_final = df[[ 'Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
df_final['current year'] = 2022
df_final['years used'] = df_final['current year']-df['Year']
df_final.drop(['Year'],axis=1, inplace=True)
df_final.drop(['current year'],axis=1, inplace=True)
df_final= pd.get_dummies(df_final, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X = df_final[['Present_Price' , 'Kms_Driven', 'years used', 'Fuel_Type_Diesel','Fuel_Type_Petrol','Transmission_Manual' ]]
x_train, x_test , y_train, y_test = train_test_split(X, df_final.Selling_Price , test_size = 0.3)
len(x_train)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```
